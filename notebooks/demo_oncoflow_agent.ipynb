This optimises for speed and reuse -> it loads the raw Xena files (used for example), harmonises sample IDs, keeps onl useful clinical/survival columns, saves smaller, clean csvs

```python
from oncoflow.loaders import load_brca_expression, load_brca_clinical, load_brca_survival

expr = load_brca_expression()
clin = load_brca_clinical()
surv = load_brca_survival()

common_samples = set(expr.columns) & set(clin.index) & set(surv.index)
expr = expr.loc[:, sorted(common_samples)]
clin = clin.loc[sorted(common_samples)]
surv = surv.loc[sorted(common_samples)]

expr.to_csv("data/brca_expression_clean.csv")
clin.to_csv("data/brca_clinical_clean.csv")
surv.to_csv("data/brca_survival_clean.csv")
```
